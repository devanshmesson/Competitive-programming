###Why we need to plot data?  
Plotting a graph visualizes the data in a graphical form which makes the data more intuitive and interpretable.

###How we can plot data in python?
Data can be plotted using a package in python which is "Matplotlib".

###What we can do after plotting data?
Data can be visualized, analyzed, some document insights can be taken out and we can transform our dataset on the basis of important and unimportant data.

###Various plots:

1.   Bar plot
2.   Histogram
3.   Scatter plot
4.   Pie plot
5.   Area plot



##Bar plot
This plot represents the comparison between the values of different category of data.

Following bar chart will be plotted to represent comparison between number of positive product reviews, negative product reviews and neutral product reviews provided as a feedback by customers. 

```python
from matplotlib import pyplot as plt  #for visualizing data
from matplotlib import style #different styles for plots
import numpy as np #for storing data efficiently

total_reviews=100

reviews=np.array(['Positive_reviews','Negative_reviews','Neutral_reviews'])
number=np.array([35,25,40])


style.use('fivethirtyeight')
plt.grid(True,color='black')
plt.title("Comparison between the number of 3 types of sentiment of reviews\n")
plt.bar(reviews,number,width=0.5,color='green',label='bar graph') #Plotting graph to canvas
plt.xlabel("Sentiment of reviews")
plt.ylabel("Number of reviews")
plt.legend() 
plt.show() #Display the plotted bar graph

print("Above bar plot shows that Number of neutral reviews were higher than the number of positive ans negative reviews!")
```

#Histogram
It represents distribution of quantitative data.

Difference between histogram and bar plot is that, Histogram can represent quantity of data in a range, but bar plot will plot each quantity seperately, which is hard to visualize in some cases.

Following is the histogram representing that majority of class scores marks in a range.


```python
style.use('fivethirtyeight')
plt.title("Range-wise distribution of marks in a class")
marks=np.array([61,95,36,33,73,30,77,56,90,22,55,58,54,80,99,25,99,55,68,29,44,49,55,51,54])
bins1=np.array([20,40,60,70,80,100])
plt.hist(marks, bins=bins1,edgecolor='black')
plt.show()
print("\nAbove histogram shows that majority of the students scored marks in the range 40-60")
```

#Scatter plot
This plot is used to visualize relationship between two variables.

Following scatter plot shows relationship between number of corona cases and number of people wearing masks.

```python
corona_cases=np.array([100,2,45,78,3,86,90,75,66,25,38,44])
wear_mask=np.array([5,100,50,20,90,7,3,15,23,75,65,55])

style.use('fivethirtyeight')
plt.title("Relationship between number of corona cases and number of people wearing masks\n\n")
plt.scatter(corona_cases,wear_mask,color='blue')
plt.xlabel('Number of Corona cases')
plt.ylabel('Number of people wearing masks')
plt.show()
print("\nAbove scatter plot shows that lesser the number of peope wearing masks, more the number of corona cases")
```

#Pie Plot
This plot is used to visualize the numerical proportion of each category among all categories.

Following Pie plot shows the proportion of different verdicts found after submitting solutions on the Online Judge

```python
Verdicts=np.array(['Wrong_Answer','Accepted_Solution','Compilation_error','Runtime_error','Time_limit_exceeded'])
Quantity=np.array([349,195,51,33,72])
style.use('fivethirtyeight')

plt.title("The proportion of different verdicts found after submitting solutions on the Online Judge\n\n")
plt.pie(Quantity,labels=Verdicts,startangle=90,shadow=True,explode=(0,0.1,0,0,0))
plt.show()
print("Above pie plot shows that, Number of Wrong answer verdicts are the highest among all other verdicts!")
```

#Area Plot
This plot is similar to line plot but the difference is that colours are filled between the x-axis and the line.

Following area plot shows rise and fall of corona cases over a period of time in months.

```python
corona_cases=np.array([100,250,310,409,510,608,550,400,300,200,100,340,490,511])
wear_mask=np.arange(14)
style.use('fivethirtyeight')

plt.title("Rise and fall of corona cases over a period of time in months\n\n")
plt.fill_between(wear_mask,corona_cases)
plt.xlabel('Time in Months')
plt.ylabel('Number of Corona cases')
plt.show()
print("\nAbove area plot shows that in first 6 months,corona cases increased drastically, then for 5 months, corona cases got decreased, Now again, number of cases are going up!")
```
